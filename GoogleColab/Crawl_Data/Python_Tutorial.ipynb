##### **Toán tử số học (Arithmetic Operators)**
Giả sử $A=10$ và $B=20$:

Toán tử | Mô tả | Ví dụ
--- | --- | ---
+ Addition | Cộng giá trị ở hai bên của toán tử | A + B = 30
– Subtraction | Trừ toán tử bên phải bằng toán tử bên trái | A – B = -10
* Multiplication | Nhân giá trị ở hai bên của toán tử | A * B = 200
/ Division | Chia giá trị bên trái với giá trị bên phải | A / B = 0.5
% Modulus | Chia giá trị bên trái với giá trị bên phải và trả về phần còn lại | B % A = 0
** Exponent	| Thực hiện tính toán hàm mũ trên các toán tử	| A**B = $10^{20}$


````
a = 21
b = 10
c = 0

c = a + b
print ( c )

c = a - b
print ( c )

c = a * b
print ( c )

c = a / b
print ( c )

c = a % b
print ( c )
a = 2
b = 3
c = a**b
print ( c )
````

##### **Toán tử so sánh (Comparison Operators)**
Các toán tử này so sánh các giá trị ở hai bên toán tử và quyết định mối quan hệ giữa chúng. Giả sử $A=10$ và $B=20$

Toán tử | Mô tả | Ví dụ
--- | --- | ---
== | Nếu giá trị của 2 toán hạng bằng nhau -> True, không bằng nhau -> False. | (A == B) -> False
!= | Nếu giá trị của 2 toán hạng không bằng nhau -> True, bằng nhau -> False. | (A != B) -> True
>	| Nếu giá trị của toán hạng bên trái lớn hơn giá trị của toán hạng bên phải -> True. | (A > B) -> False
<	| Nếu giá trị của toán hạng bên trái nhỏ hơn giá trị của toán hạng bên phải -> True. | (A < B) -> True
>= | Nếu giá trị của toán hạng bên trái lớn hơn hoặc bằng giá trị của toán hạng bên phải -> True. | (A >= B) -> False
<= | Nếu giá trị của toán hạng bên trái nhỏ hơn hoặc bằng giá trị của toán hạng bên phải -> True.. | (A <= B) -> True

````
a = 21
b = 10
c = 0

if a == b:
   print("a is equal to b")
else:
   print("a is not equal to b")

if a != b:
   print("a is not equal to b")
else:
   print("a is equal to b")

if a < b:
   print("a is less than b")
else:
   print("a is not less than b")

if a > b:
   print("a is greater than b")
else:
   print("a is not greater than b")

a = 5
b = 20
if a <= b:
   print("a is either less than or equal to b")
else:
   print("a is neither less than nor equal to b")
if a >= b:
   print("a is either greater than  or equal to b")
else:
   print("a is neither greater than  nor equal to b")
````

Trên ví dụ trên có sử dụng các lệnh điều kiện (if, else). Về cơ bản nó có nghĩa là nếu điều kiện là đúng thì thực hiện các lệnh bên trong nó, nếu không thì bỏ qua.

##### **Toán tử gán (Assignment Operators)**

Giả sử $A=10$ và $B=20$:

Toán tử | Mô tả | Ví dụ
--- | --- | ---
= |	Gán các giá trị từ toán hạng bên phải cho toán hạng bên trái. | c = a + b gán giá trị của a + b vào c
+= Add AND | Cộng toán hạng bên phải với toán hạng bên trái và gán kết quả cho toán hạng bên trái. | c += a tương đương với c = c + a
-= Subtract AND	| Trừ toán hạng bên trái với toán hạng bên phải và gán kết quả cho toán hạng bên trái.	| c -= a tương đương với c = c – a
*= Multiply AND	| Nhân toán hạng bên phải với toán hạng bên trái và gán kết quả cho toán hạng bên trái. | c *= a tương đương với c = c * a

###### **Kế thừa (Inheritance)**
Cho phép xây dựng một lớp mới dựa trên các định nghĩa của lớp đã có. Có nghĩa là lớp cha có thể chia sẽ dữ liệu và phương thức cho các lớp con. Các lớp con khỏi phải định nghĩa lại, ngoài ra có thể mở rộng các thành phần kế thừa và bổ sung thêm các thành phần mới.


````
class employee:
    num_employee=0
    raise_amount=1.04
    def __init__(self, first, last, sal):
        self.first=first
        self.last=last
        self.sal=sal
        self.email=first + '.' + last + '@company.com'
        employee.num_employee+=1

    def fullname (self):
        return '{} {}'.format(self.first, self.last)

    def apply_raise (self):
        self.sal = int(self.sal * raise_amount)

class developer(employee):
    pass

emp_1=developer('huy', 'pham', 1000000)
print(emp_1.email)
````

#### **Chuyển đổi không gian màu**
Trong OpenCV, hình ảnh không được lưu trữ ở không gian màu RGB thông thường mà là không gian màu BGR. Hàm **cvtColor()** chuyển đổi hình ảnh sang không gian màu khác.

##### **BGR -> Gray**
Công thức:

$G = 0.299 * R + 0.587 * G + 0.114 * B$

````
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
````

##### **BGR -> HSV**

![alt text](https://www.researchgate.net/profile/Siti_Nurmaini/publication/328406285/figure/fig2/AS:700381877264387@1543995460531/Simple-explorer-robot-FIGURE-3-HSV-colour-space-Red-Green-Blue-RGB-color-model-from.jpg)

Công thức:

$V = max = max(B, G, R); min = min(B, G, R)$
$S = (max - min)/max (or S = 0, if V = 0)$

$H = 30 *
  \begin{cases}
    0 + \frac{G-B}{max - min}       & \quad \text{if } max \text{ is R}\\
    2 + \frac{B-R}{max - min}       & \quad \text{if } max \text{ is G}\\
    4 + \frac{R-G}{max - min}       & \quad \text{if } max \text{ is B}
  \end{cases}
$

````
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
H = hsv[:, :, 0]
S = hsv[:, :, 1]
V = hsv[:, :, 2]
N = np.ones_like(H, dtype=np.uint8)

H_layer = cv2.merge((H, N*255, N*255))
S_layer = cv2.merge((N*180, S, N*255))
V_layer = cv2.merge((N*180, N*255, V))

H_layer = cv2.cvtColor(H_layer, cv2.COLOR_HSV2BGR)
S_layer = cv2.cvtColor(S_layer, cv2.COLOR_HSV2BGR)
V_layer = cv2.cvtColor(V_layer, cv2.COLOR_HSV2BGR)

print('H Layer')
cv2_imshow(H_layer)
print('S Layer')
cv2_imshow(S_layer)
print('V Layer')
cv2_imshow(V_layer)
````

#### **CSV File**

Một csv file (Comma Separated Values file) là một loại file văn bản thuần tuý có cấu trúc cụ thể sắp xếp dữ liệu theo dạng bảng.

Cấu trúc của file csv theo như cái tên của nó. Thông thường, các csv file sử dụng dấu phẩy để phân tách từng giá trị dữ liệu cụ thể:



```
column 1 name,column 2 name, column 3 name
first row data 1,first row data 2,first row data 3
second row data 1,second row data 2,second row data 3
...
```

Chú ý rằng mỗi phần dữ liệu đều phân tách bằng dấu phẩy. Thông thường dòng đầu tiên xác định tên cột dữ liệu. Mỗi dòng tiếp theo là dữ liệu thực.

Ngoài phân tách bằng dấu phẩy còn có các ký tự tab **\t**, dấu hai chấm **:**, dấu chấm phẩy **;**

##### **Phân tích File CSV bằng thư viện [CSV](https://docs.python.org/3/library/csv.html)**
###### **Đọc file CSV**

````
import csv

with open('csv/employee_birthday.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
            line_count += 1
    print(f'Processed {line_count} lines.')
````

###### **Đọc file CSV như một Dictionary**

````
import csv

with open('csv/employee_birthday.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        print(f'\t{row["name"]} works in the {row["department"]} department, and was born in {row["birthday month"]}.')
        line_count += 1
    print(f'Processed {line_count} lines.')
````

##### **Tách dữ liệu**
Để bóc tách dữ liệu từ một văn bản dài sử dụng thư viện beautifulSoup4.

Beautiful Soup sẽ giúp chúng ta phân tích dữ liệu HTML hay XML thành dữ liệu cây, từ đó giúp chúng ta truy xuất dữ liệu dễ dàng hơn.

````
import requests
from bs4 import BeautifulSoup

response = requests.get("https://tuoitre.vn/tin-moi-nhat.htm")
soup = BeautifulSoup(response.content, "html.parser")
print(soup)
````

##### **Phân tích dữ liệu**
Bước tiếp theo là chúng ta cần phân tích xem dữ liệu cần lấy ở đâu. Rõ ràng để lấy dữ liệu chi tiết một bài báo ta cần liên kết đến bài đó. Bật f12 lên và phân tích.

Có thể tìm thấy link bài báo ở trong thẻ $<a></a>$ và thẻ này nằm trong thẻ $h3$ có class **"title-news"**. Vậy công việc của chúng ra là lọc tất cả thẻ $h3$ có class **"title-news"** và lấy thẻ $a$ trong nó:

````
titles = soup.findAll('h3', class_='box-title-text')
print(titles)
````

Tiếp tục lấy link của tất cả các bài viết đó

````
links = [link.find('a').attrs["href"] for link in titles]
print(links)
````

Một mảng chứa các liên kết. Ở trên để tìm tất cả các thẻ $h3$ nên ta sử dụng hàm **findAll()** thõa mãn class truyền vào, nhưng ở dưới chỉ cần tìm duy nhất 1 thẻ $a$ trong $h3$ nên ta chỉ cần dùng lệnh **find("a")** và dùng **attrs["thuoc_tinh"]** để lấy thuộc tính của thẻ $a$.

#### **Lấy dữ liệu chi tiết từng bài**
Công việc tiếp theo là truy cập từng bài viết, lấy một ảnh làm đại diện và một đoạn trích ngắn.

````
for link in links:
  try:
    news = requests.get("https://tuoitre.vn" + link)
    soup = BeautifulSoup(news.content, "html.parser")
    title = soup.find("h1", class_="article-title").text
    if len(title) == 0:
      title = soup.find("h1", class_="detail-title").text
    abstract = soup.find("h2", class_="detail-sapo").text.strip()
    body = soup.find("div", class_="detail-content")
    content = '\n'.join([cnt.text.strip() for cnt in body.findChildren("p", recursive=False)])
    image = body.find("img").attrs["src"]
    print("Tiêu đề: " + title)
    print("Mô tả: " + abstract)
    print("Nội dung: " + content)
    print("Ảnh minh họa: " + image)
    print("_________________________________________________________________________")
  except:
    pass
````

Vậy là xong phần crawl dữ liệu.
Tổng hợp các dòng mã trên và viết thành một function:

````
def crawNewsData(baseUrl, url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    titles = soup.findAll('h3', class_='box-title-text')
    links = [link.find('a').attrs["href"] for link in titles]
    data = []
    for link in links:
      try:
        news = requests.get(baseUrl + link)
        soup = BeautifulSoup(news.content, "html.parser")
        title = soup.find("h1", class_="article-title").text
        if len(title) == 0:
          title = soup.find("h1", class_="detail-title").text
        abstract = soup.find("h2", class_="detail-sapo").text.strip()
        body = soup.find("div", class_="detail-content")
        content = '\n'.join([cnt.text.strip() for cnt in body.findChildren("p", recursive=False)])
        image = body.find("img").attrs["src"]
        data.append({
            "title": title,
            "abstract": abstract,
            "content": content,
            "image": image,
        })
      except:
        pass
    return data
````

````
dataNews = crawNewsData("https://tuoitre.vn", "https://tuoitre.vn/tin-moi-nhat.htm")
print(dataNews)
````
